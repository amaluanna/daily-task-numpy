1.Write a function to find the maximum of three numbers

```python
def maximum(n1,n2,n3):
    if n1>n2 and n1>n3:
        print(f'Maximum value : {n1}')
    elif n2>n1 and n2>n3:
        print(f'Maximum value : {n2}')
    else:
        print(f'Maximum value : {n3}')
n1=int(input("Enter first number : "))
n2=int(input("Enter second number : "))
n3=int(input("Enter third number : "))
maximum(n1,n2,n3)
```

2. Implement a funtion to calculate the area of a circle given the radius

```python
def area(r):
    circle_area=3.14*r**2
    print(f'Area of a circle : {circle_area}')
r=float(input("Radius : "))
area(r)
```

QUESTION 7

```python
def power(base, exponent):
    if exponent == 0:
        return 1
    elif exponent == 1:
        return base
    else:
        return base * power(base, exponent - 1)
num1=int(input('Base= '))
num2=int(input('exponent= '))
print(f"{num1}^{num2} =", power(num1, num2))
```

Question 8

```python
def digits(n):
    if n < 10:
        return n
    else:
        return (n % 10) + digits(n // 10)
num=int(input('Number= '))
print("Sum= ", digits(num))
```

QUESTION 9

```python
def gcd(a, b):
    if b == 0:
        return a
    else:
        return gcd(b, a % b)
num1=int(input('num1= '))
num2=int(input('num2= '))
print(f"GCD of {num1} and {num2} is", gcd(num1, num2))
```
